#### Forum Task 1


```python
lloyd = {
    "name": "Lloyd",
    "homework": [90.0, 97.0, 75.0, 92.0],
    "quizzes": [88.0, 40.0, 94.0],
    "tests": [75.0, 90.0],
}

alice = {
    "name": "Alice",
    "homework": [100.0, 92.0, 98.0, 100.0],
    "quizzes": [82.0, 83.0, 91.0],
    "tests": [89.0, 97.0],
}

tyler = {
    "name": "Tyler",
    "homework": [0.0, 87.0, 75.0, 22.0],
    "quizzes": [0.0, 75.0, 78.0],
    "tests": [100.0, 100.0],
}


# list of dict
students = [lloyd, alice, tyler]


for i in students:
    # print dict items
    print(
        f"Name: {i['name']}\n Homework: \t{i['homework']}\n Quizzes: \t{i['quizzes']}\n Tests: \t{i['tests']}\n"
    )


def average(nums):
    # Sum
    sum_of_nums = sum(nums)
    # Average
    total = float(sum_of_nums) / len(nums)
    return total


def get_average(student):
    # Average
    homework = average(student["homework"])
    quizzes = average(student["quizzes"])
    tests = average(student["tests"])

    # Weight
    homework = homework * 0.10
    quizzes = quizzes * 0.30
    tests = tests * 0.60

    total = sum([homework, quizzes, tests])
    return total


def get_letter_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"


print("Lloyd Letter Grade: ", get_letter_grade(get_average(lloyd)))


def get_class_average(students):
    results = []
    # just loop through student
    for i in students:
        results.append(get_average(i))

    return average(results)

# output
print("Class Grade & Letter Grade: ", get_class_average(students), get_letter_grade(get_class_average(students)))
```

#### Forum Task 3


```python
days_of_the_weeks = (
    "Sunday",
    "Monday",
    "Tuesday",
    "Wednesday",
    "Thursday",
    "Friday",
    "Saturday",
)

data = []

# i is every item in the days of the weeks
for i in days_of_the_weeks:
    store_sale = float(input(f"Enter the sales for {i}: "))

    # while still in the day
    while store_sale <= 0:
        # if condition not met
        store_sale = float(input(f"Input was invalid. Re-enter the sales for {i}: "))
    else:
        # if condition met
        data.append(store_sale)

# output
total = sum(data)
minimum = min(data)
maximum = max(data)

print(f"The total sales is: ${total:,}")
print(f"The minimum sale amount was: ${minimum:,}")
print(f"The maximum sale amount was: ${maximum:,}")
```
